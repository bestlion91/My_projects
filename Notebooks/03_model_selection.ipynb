# Model Training


```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier, AdaBoostClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from xgboost import XGBClassifier
from tabulate import tabulate
import numpy as np
```

```python
processed_data="C:/Users/aaa/OneDrive/Desktop/Modular_Project_3/Data/Processed/processed_dataset.csv"
df=pd.read_csv(processed_data)
```

```python
df.info()
```

```python
y=df['Disease_Risk']
```

```python
x=df.drop('Disease_Risk', axis=1)
```

```python
x_train,x_test, y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
x_train.shape
x_test.shape
```

# Logistic Regression

```python
lr=LogisticRegression()
```

```python
lr
```

```python
lr.fit(x_test,y_test)
```

```python
y_pred_lr=lr.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred_lr))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_lr))
print("Classification Report:\n", classification_report(y_test, y_pred_lr))
```

```python
# ['Logistic Regression', accuracy_score(y_test, y_pred_lr), confusion_matrix(y_test, y_pred_lr), classification_report(y_test, y_pred_lr)]
```

# RandomForestClassifier

```python
rf=RandomForestClassifier()
```

```python
rf
```

```python
rf.fit(x_test,y_test)
```

```python
y_pred_rf=rf.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred_rf))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_rf))
print("Classification Report:\n", classification_report(y_test, y_pred_rf))
```

```python
# ['RandomForestClassifier', accuracy_score(y_test, y_pred_rf), confusion_matrix(y_test, y_pred_rf), classification_report(y_test, y_pred_rf)]
```

# DecisionTreeClassifier

```python
dt = DecisionTreeClassifier(random_state=42)
dt
dt.fit(x_train, y_train)
y_pred_dt=dt.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred_dt))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_dt))
print("Classification Report:\n", classification_report(y_test, y_pred_dt))
```

```python
# ['DecisionTreeClassifier', accuracy_score(y_test, y_pred_dt), confusion_matrix(y_test, y_pred_dt), classification_report(y_test, y_pred_dt)]
```

# AdaBoostClassifier

```python
ad=AdaBoostClassifier()
ad
ad.fit(x_train, y_train)
y_pred_ad=ad.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred_ad))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_ad))
print("Classification Report:\n", classification_report(y_test, y_pred_ad))
```

```python
# ['AdaBoostClassifier', accuracy_score(y_test, y_pred_ad), confusion_matrix(y_test, y_pred_ad), classification_report(y_test, y_pred_ad)]
```

# KNeighborsClassifier

```python
knn=KNeighborsClassifier(n_neighbors=5)
```

```python
knn
knn.fit(x_train, y_train)
y_pred_knn=knn.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred_knn))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_knn))
print("Classification Report:\n", classification_report(y_test, y_pred_knn))
```

```python
# ['KNeighborsClassifier', accuracy_score(y_test, y_pred_knn), confusion_matrix(y_test, y_pred_knn), classification_report(y_test, y_pred_knn)]
```

# XGboost

```python
xb=XGBClassifier(eval_metric="mlogloss", random_state=42)
```

```python
xb
xb.fit(x_train, y_train)
y_pred_xb=xb.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred_xb))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_xb))
print("Classification Report:\n", classification_report(y_test, y_pred_xb))
```

```python
# ['XGBClassifier', accuracy_score(y_test, y_pred_xb), confusion_matrix(y_test, y_pred_xb), classification_report(y_test, y_pred_xb)]
```

```python
results = [
    ['Logistic Regression', accuracy_score(y_test, y_pred_lr), confusion_matrix(y_test, y_pred_lr), classification_report(y_test, y_pred_lr)],
    ['RandomForestClassifier', accuracy_score(y_test, y_pred_rf), confusion_matrix(y_test, y_pred_rf), classification_report(y_test, y_pred_rf)],
    ['GradientBoostingClassifier', accuracy_score(y_test, y_pred_gb), confusion_matrix(y_test, y_pred_gb), classification_report(y_test, y_pred_gb)],
    ['DecisionTreeClassifier', accuracy_score(y_test, y_pred_dt), confusion_matrix(y_test, y_pred_dt), classification_report(y_test, y_pred_dt)],
    ['AdaBoostClassifier', accuracy_score(y_test, y_pred_ad), confusion_matrix(y_test, y_pred_ad), classification_report(y_test, y_pred_ad)],
    ['KNeighborsClassifier', accuracy_score(y_test, y_pred_knn), confusion_matrix(y_test, y_pred_knn), classification_report(y_test, y_pred_knn)],
    ['XGBClassifier', accuracy_score(y_test, y_pred_xb), confusion_matrix(y_test, y_pred_xb), classification_report(y_test, y_pred_xb)]
]

headers = ["Model", "Accuracy", "Confusion Matrix", "Classification Report"]

table = tabulate(results, headers=headers, tablefmt='grid', floatfmt='.7f')
```

```python
print(table)
```

```python
results_dict = {model[0]: model[1] for model in results}
best_model = max(results_dict, key=results_dict.get)
print("Best Model:", best_model, "with Accuracy:", results_dict[best_model])
```
