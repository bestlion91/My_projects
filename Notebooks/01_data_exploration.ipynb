```python
# ushbu faylni yaratishdan asosiy maqsad datasetni yuklab olish va dast\labki tahlillarni qilish 
# Kerakli kutubxonalarni chaqirib olamiz
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
datasetim="C:/Users/aaa/OneDrive/Desktop/Modular_Project_3/Data/Raw/synthetic_dna_dataset.csv"
```

```python
class Dataset_tahlili:
    def __init__(self, file_path): # Datasetni yuklash
        self.df = pd.read_csv(datasetim)
    
    def show_shape_head(self, n=10): # Dataset shakli va bosh qismini ko‘rsatish
        print("Dataset shakli:", self.df.shape)
        display(self.df.head(n))
    
    def general_info(self): # Umumiy ma’lumotlar, missing values va statistikalar
        self.df.info()
        print("\nMissing values:\n", self.df.isnull().sum())
    
    def categorical_analysis(self): # Kategorik ustunlar tahlili
        categorical_cols = self.df.select_dtypes(include=["object"]).columns
        print("Kategorik ustunlar:", list(categorical_cols))
    
    def numeric_analysis(self): # Kategorik ustunlar tahlili
        numeric_cols = self.df.select_dtypes(include=[np.number]).columns
        print("Numeric ustunlar:", list(numeric_cols))

```

```python
tahlil = Dataset_tahlili(datasetim)
tahlil.show_shape_head()
tahlil.general_info()
tahlil.categorical_analysis()
tahlil.numeric_analysis()
```
